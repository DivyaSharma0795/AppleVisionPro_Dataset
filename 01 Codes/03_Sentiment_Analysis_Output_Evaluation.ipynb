```python
data_2 = data_model[
    [
        "createdAt",
        "translated_text",
        "sentiment_blob",
        "sentiment_roberta",
        "sentiment_vader",
        "sentiment_distilbert",
        "replyCount",
        "quoteCount",
        "retweetCount",
        "likeCount",
        "views",
        "bookmarkCount",
    ]
]
```

```python
data_2["createdAt"] = pd.to_datetime(data_2["createdAt"])
```

```python
import matplotlib.pyplot as plt

# Create new columns 'sentiment_category_blob', 'sentiment_category_roberta', 'sentiment_category_vader' based on sentiment scores
data_2["sentiment_category_blob"] = [
    "positive" if x > 0 else "negative" if x < 0 else "neutral"
    for x in data_2["sentiment_blob"]
]
data_2["sentiment_category_vader"] = [
    "positive" if x > 0 else "negative" if x < 0 else "neutral"
    for x in data_2["sentiment_vader"]
]
data_2["sentiment_category_distilbert"] = data_2["sentiment_distilbert"]
data_2["sentiment_category_roberta"] = data_2["sentiment_roberta"].str.lower()


# Group by date and sentiment category, and count the number of tweets in each group
grouped_blob = (
    data_2.groupby([data_2["createdAt"].dt.date, "sentiment_category_blob"])
    .size()
    .unstack(fill_value=0)
)
grouped_roberta = (
    data_2.groupby([data_2["createdAt"].dt.date, "sentiment_category_roberta"])
    .size()
    .unstack(fill_value=0)
)
grouped_vader = (
    data_2.groupby([data_2["createdAt"].dt.date, "sentiment_category_vader"])
    .size()
    .unstack(fill_value=0)
)
grouped_distilbert = (
    data_2.groupby([data_2["createdAt"].dt.date, "sentiment_category_distilbert"])
    .size()
    .unstack(fill_value=0)
)

# fig, axs = plt.subplots(4, figsize=(10, 24))

# grouped_blob.plot(kind="bar", stacked=True, ax=axs[0])
# axs[0].set_title("Blob Model")
# axs[0].set_xlabel("Date")
# axs[0].set_ylabel("Number of Tweets")


# grouped_vader.plot(kind="bar", stacked=True, ax=axs[1])
# axs[1].set_title("Vader Model")
# axs[1].set_xlabel("Date")
# axs[1].set_ylabel("Number of Tweets")

# grouped_distilbert.plot(kind="bar", stacked=True, ax=axs[2])
# axs[2].set_title("Distilbert Model")
# axs[2].set_xlabel("Date")
# axs[2].set_ylabel("Number of Tweets")

# grouped_roberta.plot(kind="bar", stacked=True, ax=axs[3])
# axs[3].set_title("Roberta Model")
# axs[3].set_xlabel("Date")
# axs[3].set_ylabel("Number of Tweets")
# plt.tight_layout()
# plt.show()
```

```python
# Blob Model
plt.figure(figsize=(10, 6))
grouped_blob.plot(kind="bar", stacked=True)
plt.title("TextBlob Model")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(range(0, len(grouped_blob.index), 10), grouped_blob.index[::10], rotation=45)
plt.show()

# Vader Model
plt.figure(figsize=(10, 6))
grouped_vader.plot(kind="bar", stacked=True)
plt.title("VADER Model")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(
    range(0, len(grouped_vader.index), 10), grouped_vader.index[::10], rotation=45
)
plt.show()

# Distilbert Model
plt.figure(figsize=(10, 6))
grouped_distilbert.plot(kind="bar", stacked=True)
plt.title("DistilBERT Model")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(
    range(0, len(grouped_distilbert.index), 10),
    grouped_distilbert.index[::10],
    rotation=45,
)
plt.show()

# Roberta Model
plt.figure(figsize=(10, 6))
grouped_roberta.plot(kind="bar", stacked=True)
plt.title("RoBERTa Model")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(
    range(0, len(grouped_roberta.index), 10), grouped_roberta.index[::10], rotation=45
)
plt.show()
```

```python
# Normalize the data by row
grouped_blob_percentage = grouped_blob.div(grouped_blob.sum(axis=1), axis=0)
grouped_vader_percentage = grouped_vader.div(grouped_vader.sum(axis=1), axis=0)
grouped_distilbert_percentage = grouped_distilbert.div(
    grouped_distilbert.sum(axis=1), axis=0
)
grouped_roberta_percentage = grouped_roberta.div(grouped_roberta.sum(axis=1), axis=0)

# Blob Model
plt.figure(figsize=(10, 6))
grouped_blob_percentage.plot(kind="bar", stacked=True)
plt.title("TextBlob Model")
plt.xlabel("Date")
plt.ylabel("Percentage of Tweets")
plt.xticks(range(0, len(grouped_blob.index), 10), grouped_blob.index[::10], rotation=45)
plt.show()

# Vader Model
plt.figure(figsize=(10, 6))
grouped_vader_percentage.plot(kind="bar", stacked=True)
plt.title("VADER Model")
plt.xlabel("Date")
plt.ylabel("Percentage of Tweets")
plt.xticks(
    range(0, len(grouped_vader.index), 10), grouped_vader.index[::10], rotation=45
)
plt.show()

# Distilbert Model
plt.figure(figsize=(10, 6))
grouped_distilbert_percentage.plot(kind="bar", stacked=True)
plt.title("DistilBERT Model")
plt.xlabel("Date")
plt.ylabel("Percentage of Tweets")
plt.xticks(
    range(0, len(grouped_distilbert.index), 10),
    grouped_distilbert.index[::10],
    rotation=45,
)
plt.show()

# Roberta Model
plt.figure(figsize=(10, 6))
grouped_roberta_percentage.plot(kind="bar", stacked=True)
plt.title("RoBERTa Model")
plt.xlabel("Date")
plt.ylabel("Percentage of Tweets")
plt.xticks(
    range(0, len(grouped_roberta.index), 10), grouped_roberta.index[::10], rotation=45
)
plt.show()
```

```python
# Remove the '12-31-2023' date column
for model in models:
    if "2023-12-31" in model.index:
        model.drop(pd.Timestamp("2023-12-31"), inplace=True)

# Convert the index to a DatetimeIndex
grouped_blob.index = pd.to_datetime(grouped_blob.index)
grouped_vader.index = pd.to_datetime(grouped_vader.index)
grouped_distilbert.index = pd.to_datetime(grouped_distilbert.index)
grouped_roberta.index = pd.to_datetime(grouped_roberta.index)

# Resample the data to a monthly frequency
grouped_blob_monthly = grouped_blob.resample("M").sum()
grouped_vader_monthly = grouped_vader.resample("M").sum()
grouped_distilbert_monthly = grouped_distilbert.resample("M").sum()
grouped_roberta_monthly = grouped_roberta.resample("M").sum()

# Normalize the data by row
grouped_blob_percentage_monthly = grouped_blob_monthly.div(
    grouped_blob_monthly.sum(axis=1), axis=0
)
grouped_vader_percentage_monthly = grouped_vader_monthly.div(
    grouped_vader_monthly.sum(axis=1), axis=0
)
grouped_distilbert_percentage_monthly = grouped_distilbert_monthly.div(
    grouped_distilbert_monthly.sum(axis=1), axis=0
)
grouped_roberta_percentage_monthly = grouped_roberta_monthly.div(
    grouped_roberta_monthly.sum(axis=1), axis=0
)

# Plot the data
models = [
    grouped_blob_percentage_monthly,
    grouped_vader_percentage_monthly,
    grouped_distilbert_percentage_monthly,
    grouped_roberta_percentage_monthly,
]
titles = ["TextBlob Model", "VADER Model", "DistilBERT Model", "RoBERTa Model"]

# Filter the data to only include dates in 2024
models = [
    model.loc[pd.Timestamp("2024-01-01") : pd.Timestamp("2024-12-31")]
    for model in models
]

# Plot the data with data labels
for model, title in zip(models, titles):
    plt.figure(figsize=(10, 6))
    ax = model.plot(kind="bar", stacked=True)
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Percentage of Tweets")
    plt.xticks(rotation=45)

    # Add data labels
    for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy()
        ax.text(
            x + width / 2,
            y + height / 2,
            "{:.0f} %".format(height * 100),
            horizontalalignment="center",
            verticalalignment="center",
        )
    plt.show()
```
